## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

An SQL database is a relational database.  Every entry has every bit of data - in other words, there's a value for every column and a column for every value.  NoSQL databases aren't relational.  Data can be stored differently and there aren't necessarily the same columns for each entry that's being stored.





## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Storing conversations could be better done in a NoSQL database since each conversation has a different length (and so an unknown number of columns).  Another example is a set of data where the relationships are important to keep track of as well as the info for each datapoint (like on Facebook).




## Question 3

How does MongoDB work? How is it different from a relational database?

_______________________________________________________________________________________________________________________________

MongoDB is a document database.  It stores information in a form similar to a dictionary, with a key-value pair. It's flexible and easy to create documents and store data but it's difficult to query/find data.






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
Fault tolerance is a way to ensure that all of the data is present for a query (or that the user at least knows some data is missing).  There are a lot of ways this could be done but a common way is to have backups of various parts of the data distributed amongst the servers where the data is stored.  That way, if a server is down or unavailable, the data can be pulled from somewhere else.







## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MapReduce is a way of retrieving data from several servers at once.  Each server is queried at the same time with the same query (this is the Map step) and then the data is aggregated after it's returned (this is the Reduce step).

Hadoop generally has giant datasets which would take forever to query if all of the data was stored on a single server. Instead, the data is split up on several different servers and queried using the MapReduce process.
